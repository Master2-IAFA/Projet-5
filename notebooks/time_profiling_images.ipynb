```python
import k3d
import datetime
import numpy as np
import matplotlib.pyplot as plt
import h5py
import sys
import tqdm
```

```python
dts = []
t_old = None
with open('/logs/1.err') as f:
    for line in f:
        date, pose = line.strip().split('  ')
        print(pose)
        t_now = datetime.datetime.strptime(date, '%d.%m.%Y %H:%M:%S.%f')
        if None is not t_old:
            dts.append((t_now - t_old).total_seconds() * 1000)
            print(pose, dts[-1])
        t_old = t_now
```

```python
f = plt.figure(figsize=(12, 8))
_ = plt.hist(dts, bins=100)
plt.gca().set_yscale('log')
```
